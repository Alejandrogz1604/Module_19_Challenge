```python
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

df_market_data.head(10)
```

```python
df_market_data.describe()
```

```python
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

```python
scaled_data = StandardScaler().fit_transform(df_market_data)
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data, columns = df_market_data.columns)

coin_index = df_market_data.index

scaled_df.set_index(coin_index, inplace = True)

scaled_df.head()
```

```python
k_values = list(range(1,12))
```

```python
inertia_values = []
```

```python
for i in k_values:
    k_model = KMeans(n_clusters = i, random_state = 1)
    k_model.fit(scaled_df)
    inertia_values.append(k_model.inertia_)
    
```

```python
elbow_data = {"k": k_values, "inertia": inertia_values}
elbow_data_df = pd.DataFrame(elbow_data)
elbow_data_df.head()
```

```python
elbow_curve = elbow_data_df.hvplot.line(
    x = "k", 
    y = "inertia", 
    title = "Elbow Curve",
    xlabel = "Number of Clusters",
    ylabel = "Inertia",
    xticks = k_values
)

elbow_curve
```

```python
print("Based on our elbow curve graph using the scaled data, we can begin to see an elbow form around 4 clusters.")
print("\n")
print("Visually, it appears that after four clusters we start to see diminishing returns in further clustering our data. 4 Clusters appear to be the optimal k value.")
```

```python
KMeans_model = KMeans(n_clusters = 4, random_state = 1)
KMeans_model
```

```python
KMeans_model.fit(scaled_df)
```

```python
cluster_prediction = KMeans_model.predict(scaled_df)
cluster_prediction
```

```python
copy_scaled_df = scaled_df.copy()
copy_scaled_df["Predicted Clusters"] = cluster_prediction
copy_scaled_df.head()
```

```python
scatter_plot = copy_scaled_df.hvplot.scatter(
    x = "price_change_percentage_24h",
    y = "price_change_percentage_7d",
    by = "Predicted Clusters",
    hover_cols = "coin_id",
    colorbar = True,
    title = "Cryptocurrency Price Variations")
scatter_plot
```

```python
pca_model = PCA(n_components=3)
```

```python
pca_data = pca_model.fit_transform(scaled_df)
```

```python
pca_df = pd.DataFrame(pca_data, columns = ["PC 1", "PC 2", "PC 3"])

pca_df.head()
```

```python
exp_variance = pca_model.explained_variance_ratio_
for i, ratio in enumerate(exp_variance):
    print(f"The explained variance ratio for component {i + 1} is {ratio}")
    print("\n")

print(f"The total explained variance is {exp_variance.sum()} or 89.50%")
```

```python
pca_df_2 = pd.DataFrame(pca_data, columns = ["PCA 1", "PCA 2", "PCA 3"])
pca_df_2.set_index(coin_index, inplace = True)
pca_df_2.head()
```

```python
k_pca_values = list(range(1,12))
```

```python
inertia_pca_values = []

for i in k_pca_values:
    k_pca_model = KMeans(n_clusters = i, random_state = 1)
    k_pca_model.fit(pca_df_2)
    inertia_pca_values.append(k_pca_model.inertia_)
    
```

```python
elbow_pca_data = {"k": k_pca_values, "inertia": inertia_pca_values}
elbow_pca_data_df = pd.DataFrame(elbow_pca_data)
elbow_pca_data_df.head()
```

```python
elbow_pca_curve = elbow_pca_data_df.hvplot.line(
    x = "k", 
    y = "inertia", 
    title = "PCA Elbow Curve",
    xlabel = "Number of Clusters",
    ylabel = "Inertia",
    xticks = k_values)

elbow_pca_curve
```

```python
print("The best value for K when using our PCA data is 4, which coincides with our K value from our original dataset.")
```

```python
KMeans_model_2 = KMeans(n_clusters = 4, random_state = 1)
KMeans_model_2
```

```python
KMeans_model_2.fit(pca_df_2)
```

```python
n_clusters = KMeans_model_2.predict(pca_df_2)
n_clusters
```

```python
pca_clusters_df = pca_df_2.copy()
pca_clusters_df["Predicted Clusters"] = n_clusters
pca_clusters_df.head()
```

```python
scatter_plot_2 = pca_clusters_df.hvplot.scatter(
    x = "PCA 1",
    y = "PCA 2",
    by = "Predicted Clusters",
    title = "Cryptocurrency Price Variations Using PCA.",
    hover_cols = ["coin_id"],
    xlabel= "PCA 1",
    ylabel= "PCA 2",
    colorbar= True
)

scatter_plot_2
```

```python
elbow_curve_comparison = elbow_curve + elbow_pca_curve
elbow_curve_comparison
```

```python
cluster_analysis_comparison = scatter_plot + scatter_plot_2
cluster_analysis_comparison
```

```python
print("While both methods lead to the same optimal K value of 4, the elbow curve in our PCA analysis was much more clearly defined.")
print("\n")
print("This is due to the fact that by using PCA analysis, we have been able to eliminate variables that do not have as much effect on our model.")
print("Therefore, this leads to a clearer elbow curve and more well defined clusters.")
```
